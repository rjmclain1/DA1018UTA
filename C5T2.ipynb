```python
# Need to factor Education.  4 'Other' categories is going to mess things up.
credit['EDUCATION'] = pd.Categorical(credit.EDUCATION)
credit["EDUCATION"].cat.categories = ["Other0","Graduate School","University","High School","Other4","Other5","Other6"]
```

```python
# So recode to the rescue.
credit.groupby('EDUCATION')['EDUCATION'].count()
```

```python
def recode_education(series):
    if (series == 'Other0') or (series == 'Other4') or (series == 'Other5') or (series == 'Other6'):
        return 'Other'
    else:
        return series
```

```python
credit['EDUCATION'] = credit['EDUCATION'].apply(recode_education)
```

```python
# Now, what does our data look like?
credit.groupby('EDUCATION')['EDUCATION'].count()
```
